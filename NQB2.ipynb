## Sentiment Analysis

```python
import nltk
nltk.download('vader_lexicon')
```

```python
from nltk.sentiment import SentimentIntensityAnalyzer

def senderSentimentVADER(sender_words):
    sia = SentimentIntensityAnalyzer()
    sentiment_scores = {}

    for sender, words in sender_words.items():
        text = ' '.join(words)
        sentiment = sia.polarity_scores(text)
        sentiment_scores[sender] = sentiment['compound']

    return sentiment_scores
```

```python
sender_words = waf.allWordsBySender(df)
sender_messages = df.groupby('sender')['message'].apply(list).to_dict()
sentiment_scores = senderSentimentVADER(sender_messages)

sender_messages
```

```python
pd.Series(sentiment_scores).sort_values(ascending=False).head(20)
```

```python
import pandas as pd

def sentimentAnalyse(df):
    sia = SentimentIntensityAnalyzer()

    df['message'] = df['message'].astype(str)

    sentiment_df = df['message'].apply(sia.polarity_scores).apply(pd.Series)
    df = pd.concat([df, sentiment_df], axis=1)

    aggregated = df.groupby('sender')[['neg', 'neu', 'pos', 'compound']].mean()

    return aggregated.sort_values('compound', ascending=False)

sentimentAnalyse(df)
```
